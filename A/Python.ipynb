```python
def menor_dif_tres(a,b,c): #criando nossa propria fun√ß√£o
    """
    ---

    calcula a dif absoluta entre tres numeros e retorna a menor dif entre eles
    ex:
    menor_dif_tres(30,20,24)
    4

    ---
    """
    d1 = abs(a-b)
    d2 = abs(b-c)
    d3 = abs(c-a)
    return min(d1,d2,d3)

help(menor_dif_tres) # mostra a defini√ß√£o entre "" da nossa fun√ß√£o
print(menor_dif_tres(30,20,24))

print(1,2,3)
print (1,2,3, sep= ' ')
print (1,2,3, sep= ',')

def dividapordois(d,n=2):
    "-> divide por dois, e divide por n se for posto uma segunda entrada"
    return d//2

help(dividapordois)

print(dividapordois(50))
print(dividapordois(60,3))
```
